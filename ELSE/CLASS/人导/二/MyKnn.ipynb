```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris
```

```python
iris.data
# iris.data.shape
# iris.target
# iris.target.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.3,random_state=33)
X_train.shape
y_train
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)  # fit_transform()是先拟合数据，然后转化它将其转化为标准形式
X_test = ss.transform(X_test)  # transform()是通过找中心和缩放等实现标准化
X_train
```

```python

from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier()#KNeighborsClassifier 用于分类
knc.fit(X_train,y_train)#训练模型
```

```python
y_predict=knc.predict(X_test)#预测

print('The accuracy of K-Nearest Neighbor Classifier is',knc.score(X_test,y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict,target_names=iris.target_names))
```
