# K-means聚类算法在Iris数据集上的应用


## 导入库

```python
#!/usr/bin/env python
# coding: utf-8

import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
# from matplotlib.colors import ListedColormap

from sklearn.datasets import load_iris
# cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])  # 浅色调颜色映射
```

## 加载Iris数据集

```python
iris = load_iris()
X = iris.data[:]  # 特征数据
target = iris.target  # 目标变量
```

## 使用K-means聚类算法进行聚类，聚类数量为3

```python
estimator = KMeans(n_clusters=3)
estimator.fit(X)  # 进行聚类
label_pred = estimator.labels_  # 获取聚类标签
```

## 根据聚类标签提取每个簇的样本数据

```python
k = 0
x0 = X[label_pred == 0]
x1 = X[label_pred == 1]
x2 = X[label_pred == 2]
```

## 遍历特征的组合

```python
# 创建一个2行3列的画布
fig, axs = plt.subplots(2, 3, figsize=(15, 12))
for i in range(4):
    for j in range(i + 1, 4):
        k += 1
        samples = X[:, [i, j]]

        # 绘制每个簇的散点图，使用不同颜色和标记区分簇
        plt.scatter(x0[:, i], x0[:, j], c="red", marker='o', label='label0')
        plt.scatter(x1[:, i], x1[:, j], c="green", marker='*', label='label1')
        plt.scatter(x2[:, i], x2[:, j], c="blue", marker='+', label='label2')

        # 在指定的子图中进行绘制
        plt.subplot(2, 3, k)
        axs[(k - 1) // 3, k % 3].set_xlabel(iris.feature_names[i])
        axs[(k - 1) // 3, k % 3].set_ylabel(iris.feature_names[j])
        plt.suptitle('Kmeans(3)', fontsize=50, fontweight='bold')

plt.show()
```

## 使用K-means聚类算法进行聚类，聚类数量为4

```python

estimator = KMeans(n_clusters=4)
estimator.fit(X)  # 进行聚类
label_pred = estimator.labels_  # 获取聚类标签
k = 0
# 根据聚类标签提取每个簇的样本数据
x0 = X[label_pred == 0]
x1 = X[label_pred == 1]
x2 = X[label_pred == 2]
x3 = X[label_pred == 3]

# 创建一个2行3列的画布
fig, axs = plt.subplots(2, 3, figsize=(15, 12))

# 遍历特征的组合
for i in range(4):
    for j in range(i + 1, 4):
        k += 1

        # 绘制每个簇的散点图，使用不同颜色和标记区分簇
        plt.scatter(x0[:, i], x0[:, j], c="red", marker='o', label='label0')
        plt.scatter(x1[:, i], x1[:, j], c="green", marker='*', label='label1')
        plt.scatter(x2[:, i], x2[:, j], c="blue", marker='+', label='label2')
        plt.scatter(x3[:, i], x3[:, j], c="yellow", marker='x', label='label3')

        # 在指定的子图中进行绘制
        plt.subplot(2, 3, k)
        axs[(k - 1) // 3, k % 3].set_xlabel(iris.feature_names[i])
        axs[(k - 1) // 3, k % 3].set_ylabel(iris.feature_names[j])
        plt.suptitle('Kmeans(4)', fontsize=50, fontweight='bold')

plt.show()
```

## 使用K-means聚类算法进行聚类，聚类数量为5

```python

estimator = KMeans(n_clusters=5)
estimator.fit(X)  # 进行聚类
label_pred = estimator.labels_  # 获取聚类标签
k = 0
# 根据聚类标签提取每个簇的样本数据
x0 = X[label_pred == 0]
x1 = X[label_pred == 1]
x2 = X[label_pred == 2]
x3 = X[label_pred == 3]
x4 = X[label_pred == 4]

# 创建一个2行3列的画布
fig, axs = plt.subplots(2, 3, figsize=(15, 12))

# 遍历特征的组合
for i in range(4):
    for j in range(i + 1, 4):
        k += 1

        # 绘制每个簇的散点图，使用不同颜色和标记区分簇
        plt.scatter(x0[:, i], x0[:, j], c="red", marker='o', label='label0')
        plt.scatter(x1[:, i], x1[:, j], c="green", marker='*', label='label1')
        plt.scatter(x2[:, i], x2[:, j], c="blue", marker='+', label='label2')
        plt.scatter(x3[:, i], x3[:, j], c="yellow", marker='x', label='label3')
        plt.scatter(x4[:, i], x4[:, j], c="black", marker='^', label='label4')

        # 在指定的子图中进行绘制
        plt.subplot(2, 3, k)
        axs[(k - 1) // 3, k % 3].set_xlabel(iris.feature_names[i])
        axs[(k - 1) // 3, k % 3].set_ylabel(iris.feature_names[j])
        plt.suptitle('Kmeans(5)', fontsize=50, fontweight='bold')

plt.show()
```
