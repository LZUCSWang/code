```python
#code-4-1.py

# 导入所需库和模块
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt

# 加载Boston房价数据集
dataset = load_boston()
x_data = dataset.data  # 导入所有特征变量
y_data = dataset.target  # 导入目标值（房价）
name_data = dataset.feature_names  # 导入特征

# 创建一个子图，绘制房间数与房价之间的线性关系
plt.subplot(1, 1, 1)
plt.scatter(x_data[:, 5], y_data, s=20)  # 绘制散点图，x轴为房间数，y轴为房价
plt.plot([5., 8.5], [8., 48.], color='red', lw=3)  # 绘制一条红色的直线表示线性关系
plt.xlabel('Room Number')  # 设置x轴标签为房间数
plt.ylabel('Price')  # 设置y轴标签为房价
plt.title(name_data[5])  # 设置图标题为特征名称中的第5个特征

# 打印前20个样本的房间数和房价
for i in range(0, 20):
    print(x_data[i, 5], y_data[i])

plt.show()  # 显示图形
```
